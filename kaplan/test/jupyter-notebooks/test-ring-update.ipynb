```python
from random import choice
```

```python
# this is the code that generates possible slots for new pmems
# to make sure it works, the slots should be on either side of the
# parent +/- pmem_dist
def get_slots(num_slots, pmem_dist, parent_index):
    possible_slots = []
    # first check if the range loops round the ring
    if parent_index + pmem_dist >= num_slots:
        possible_slots.extend(range(parent_index, num_slots))
        overflow = parent_index + pmem_dist - num_slots
        possible_slots.extend(range(overflow+1))
    else:
        possible_slots.extend(range(parent_index, parent_index+pmem_dist+1))
    # then check if loops around ring backwards
    # essentially checking if parent_index - pmem_dist is negative
    if parent_index < pmem_dist:
        possible_slots.extend(range(0, parent_index))
        backflow = num_slots - (pmem_dist - parent_index)
        possible_slots.extend(range(backflow, num_slots))
    else:
        possible_slots.extend(range(parent_index-pmem_dist, parent_index))

    assert len(possible_slots) == 2*pmem_dist+1
    print(possible_slots)
    print(set(possible_slots))
    print(choice(list(set(possible_slots))))
```

```python
for i in range(20):
    print(i)
    get_slots(20, 3, i)
```
