```python
from vetee.xyz import Xyz
from vetee.structure import Structure
```

```python
trans = Structure("cid", "5371758")
cis = Structure("cid", "5463122")
```

```python
trans_xyz = Xyz()
for atom in trans.coords:
    trans_xyz.coords.append([atom[0], atom[1], atom[2], atom[3]])
trans_xyz.num_atoms = len(trans_xyz.coords)
trans_xyz.write_xyz("trans.xyz")
```

```python
cis_xyz = Xyz()
for atom in cis.coords:
    cis_xyz.coords.append([atom[0], atom[1], atom[2], atom[3]])
cis_xyz.num_atoms = len(cis_xyz.coords)
cis_xyz.write_xyz("cis.xyz")
```
